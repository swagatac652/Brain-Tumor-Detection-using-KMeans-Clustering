```python
import cv2
from matplotlib import pyplot as plt

img=cv2.imread('test_images/0 (239).jpg',0) #the parameter 0 is given so that the image is loaded as a gray scale image
print(img.shape)
height,width=img.shape
plt.imshow(img,cmap='gray')

#consider 512*512 pixels as features
#apply 262144 dataset into kmeans clustering
#algo groups pixels into k clusters depending on the colour
```

```python
#flatten the image
imgFlatten=img.reshape(height*width,1)
#shape of the flattened image
print(imgFlatten.shape)
```

```python
from sklearn.cluster import KMeans

model=KMeans(n_clusters=6) #6 clusters
model.fit(imgFlatten) #passing only features ie unsupervised learning
labels=model.labels_
print(labels.shape)
labels2D=labels.reshape(height,width)
print(labels2D.shape)
```

```python
print(img[200:210,200:210].T) #.T for transpose
print('----------------------------------------------------')
print(labels2D[200:210,200:210])
```

```python
plt.imshow(labels2D)
plt.colorbar()
plt.show()
```
